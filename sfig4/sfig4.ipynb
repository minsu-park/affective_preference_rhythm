```r
library(ggplot2)
```

```r
tmp <- read.csv("../data/hour_dow_activity.csv", header=TRUE)
tmp$dow <- factor(tmp$dow,
                      levels = c("Mon","Tue", "Wed", "Thu", "Fri", "Sat", "Sun"),
                      ordered = TRUE)

p <- ggplot(tmp,aes(x=hours,y=dow)) +
  geom_tile() +
  coord_equal() +
  geom_tile(aes(fill=scaled_freq), colour="white") +
  scale_fill_gradient(low="white", high="gray20") + 
  xlab("Time of Day") + 
  ylab("Day of Week") + 
  labs(fill="Scaled Frequency\n") +
  scale_x_continuous(limits=c(-0.5,23.5), expand=c(0,0), breaks=0:23) +
  theme_bw(base_size=24)

p

pdf(file="../output/sfig4.pdf", width=23.6, height=9.1)
print(p)
dev.off()
```
